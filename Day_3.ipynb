## List Assignment: (Party)



image.png

**Scenario:** Imagine you're having a party, and you want to maintain a list of guests. Write a program that uses Python list to add guests, remove guests, and check who's on the list.

````python
def display_menu():
    """Display the menu options."""
    print("\nParty Guest List Manager")
    print("1. Add a guest")
    print("2. Remove a guest")
    print("3. Check if a guest is on the list")
    print("4. View all guests")
    print("5. Exit")

def add_guest(guest_list):
    """Add a guest to the list."""
    guest_name = input("Enter the name of the guest to add: ").strip()
    if guest_name in guest_list:
        print(f"{guest_name} is already on the guest list.")
    else:
        guest_list.append(guest_name)
        print(f"{guest_name} has been added to the guest list.")

def remove_guest(guest_list):
    """Remove a guest from the list."""
    guest_name = input("Enter the name of the guest to remove: ").strip()
    if guest_name in guest_list:
        guest_list.remove(guest_name)
        print(f"{guest_name} has been removed from the guest list.")
    else:
        print(f"{guest_name} is not on the guest list.")

def check_guest(guest_list):
    """Check if a guest is on the list."""
    guest_name = input("Enter the name of the guest to check: ").strip()
    if guest_name in guest_list:
        print(f"{guest_name} is on the guest list.")
    else:
        print(f"{guest_name} is not on the guest list.")

def view_all_guests(guest_list):
    """View all guests on the list."""
    if guest_list:
        print("\nCurrent Guest List:")
        for i, guest in enumerate(guest_list, start=1):
            print(f"{i}. {guest}")
    else:
        print("The guest list is currently empty.")

def main():
    """Main program to manage the guest list."""
    guest_list = []  # Initialize an empty guest list
    while True:
        display_menu()
        choice = input("Choose an option (1-5): ").strip()
        
        if choice == "1":
            add_guest(guest_list)
        elif choice == "2":
            remove_guest(guest_list)
        elif choice == "3":
            check_guest(guest_list)
        elif choice == "4":
            view_all_guests(guest_list)
        elif choice == "5":
            print("Exiting the Guest List Manager. Have a great party!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

# Run the program
main()
````

Dictionaries are a way to store information that is connected in some way.

Dictionaries are used to store data values in key:value pairs.

### General Syntax


````python
# Creating a dictionary
student = {
    "name": "John",
    "age": 20,
    "course": "Computer Science"
}
print(student)
````

### **Characteristics of Dictionaries**

* Keys are always unique, and there cannot be any duplicates.
* There is no index in the dictionary, meaning they are not ordered. The key is the default iterator and is used to retrieve the value.
* Dictionaries are mutable (changeable)

````python
#examples to demonstratte these properties

fruits_dict = {"apple": 1 , "oranges": 2, "mangos":3}

fruits_dict['apple'] = 4

print(fruits_dict)
````

## Dictionary Methods
Dictionaries offer various methods that help in accessing, modifying, and managing the key:value pairs.

### **Common operations with dictionaries**
There are a few common things you will want to do with dictionaries.
* Adding new key-value pairs,
* Modifying information in the dictionary,
* Removing items from dictionaries.

**Adding new key-value pairs**

````python
# Create an empty dictionary.
african_countries = {}

print(african_countries)


# Fill the dictionary with country-capital pairs.
african_countries['Nigeria'] = 'Abuja'


print(african_countries)
````

````python
african_countries['Kenya'] = 'Nairobi'
african_countries['Egypt'] = 'Cairo'
african_countries['South Africa'] = 'Pretoria'
african_countries['Ghana'] = 'Accra'

# Print out the countries and their capitals.
for country, capital in african_countries.items():
    print("\nCountry: %s" % country)
    print("Capital: %s" % capital)
````

**Removing key-value pairs**

### Dictionaries and Loops

Dictionaries can be traversed using loops, which allows you to iterate over each key:value pair. We can achieve this in 3 ways:
* You can loop through all key-value pairs;
* You can loop through the keys, and pull out the values for any keys that you care about;
* You can loop through the values.


**Looping through all key-value pairs**

````python
# Iterating through a dictionary using a for loop

for country, capital in african_countries.items():
  print(country)
  print(capital)
````

**Looping through all keys in a dictionary**

````python
my_dict = {'key_1': 'value_1',
    'key_2': 'value_2',
    'key_3': 'value_3',
    }

for key in my_dict.keys():
    print('Key: %s' % key)

#for key in my_dict:
#    print('Key: %s' % key)
````

**Looping through all values in a dictionary**

````python
my_dict = {'key_1': 'value_1',
    'key_2': 'value_2',
    'key_3': 'value_3',
    }

for value in my_dict.values():
    print('Value: %s' % value)
````

# Tuples

Tuples are basically lists that can never be changed.

Tuples in Python are ordered, immutable collections. This means that once a tuple is created, its content cannot be altered. Tuples can contain mixed data types and are similar to lists, but they have certain advantages in specific contexts due to their immutability.


````python
# Creating a tuple
sample_tuple = (1, 2, "Python", 3.5)
print(sample_tuple)
````

````python

# convert from a list
tuple(['Veery', 'Eastern Meadowlark', 10, 25, True])
````

````python

# blank tuple
tuple()
````

**Accessing elements**

````python
countries = list(african_countries.keys())

print(countries)
````

````python
c_tup = tuple(countries)
print(c_tup)
````

````python
c_tup[2]
````

**Slicing**

````python
c_tup[1:4]
````

### **Tuples and Lists**

````python
# Converting between lists and tuples
sample_list = [1, 2, 3, 4]
converted_tuple = tuple(sample_list)
print(converted_tuple)

converted_list = list(converted_tuple)
print(converted_list)
````

### **Tuples as Return Values**
* Tuples are often used in functions to return multiple values simultaneously

````python
# Function that returns a tuple
def min_max(numbers):
    return (min(numbers), max(numbers))

result = min_max([3, 1, 4, 2, 5])
print(result)  # Outputs: (1, 5)
````

### **(Optional) Dictionaries and Tuples**

````python
# Extracting keys and values as tuples and sorting
word_freq = {"Python": 3, "is": 3, "great": 1}
sorted_by_key = sorted(word_freq.items())
print(sorted_by_key)

# Sorting by values
sorted_by_value = sorted(word_freq.items(), key=lambda item: item[1], reverse=True)
print(sorted_by_value)
````

## Assignment (Advanced Party Guest Manager)


**Objective:**
Practice using Python dictionaries and tuples by building a guest management system for a party.


## **Steps To Complete**

1. Guest List Initialization:

Create an empty dictionary called guests to store guest information.
Each guest will have a name, age, and email.
2. Adding Guests:

Add the following guests to your dictionary using tuples:
"Alice", 28, "alice@email.com"
"Bob", 35, "bob@email.com"
"Charlie", 30, "charlie@email.com"
Use the name as the dictionary key, and the tuple (age, email) as the value.
3. Updating the Guest List:

Add a new guest: "David", 22, "david@email.com".
Remove "Bob" from the list.
4. Guest Information Retriever:

Create a function get_guest_info(guest_name) that:
Checks if the guest is on the list.
If yes, returns: "{guest_name} (Age: {age}) is coming to the party! Email: {email}"
Otherwise, returns: "{guest_name} is not on the guest list."
5. Guest Counter:

Display the total number of guests currently on the list.

**Extensions:**

Allow users to input their own guests, with the required details.
Display guests in order of age.
Handle the scenario where someone tries to add an existing guest. Prompt to update details or skip.

````python
# Initialize the guest dictionary
guests = {}

# Adding predefined guests to the dictionary
guests["Alice"] = (28, "alice@email.com")
guests["Bob"] = (35, "bob@email.com")
guests["Charlie"] = (30, "charlie@email.com")

# Function to add a new guest
def add_guest(name, age, email):
    if name in guests:
        print(f"{name} is already on the guest list.")
        update = input("Do you want to update their details? (yes/no): ").strip().lower()
        if update == "yes":
            guests[name] = (age, email)
            print(f"{name}'s details have been updated.")
        else:
            print(f"Skipped updating {name}'s details.")
    else:
        guests[name] = (age, email)
        print(f"{name} has been added to the guest list.")

# Function to remove a guest
def remove_guest(name):
    if name in guests:
        del guests[name]
        print(f"{name} has been removed from the guest list.")
    else:
        print(f"{name} is not on the guest list.")

# Function to retrieve guest information
def get_guest_info(guest_name):
    if guest_name in guests:
        age, email = guests[guest_name]
        return f"{guest_name} (Age: {age}) is coming to the party! Email: {email}"
    else:
        return f"{guest_name} is not on the guest list."

# Function to display the total number of guests
def guest_count():
    return len(guests)

# Function to display guests sorted by age
def display_guests_by_age():
    sorted_guests = sorted(guests.items(), key=lambda g: g[1][0])
    print("\nGuests Sorted by Age:")
    for name, (age, email) in sorted_guests:
        print(f"{name} (Age: {age}, Email: {email})")

# Adding a new guest "David" and removing "Bob"
add_guest("David", 22, "david@email.com")
remove_guest("Bob")

# Interactive menu for the guest manager
def guest_manager():
    while True:
        print("\nAdvanced Party Guest Manager")
        print("1. Add a guest")
        print("2. Remove a guest")
        print("3. Get guest information")
        print("4. View guests sorted by age")
        print("5. Display total number of guests")
        print("6. Exit")
        
        choice = input("Choose an option (1-6): ").strip()
        
        if choice == "1":
            name = input("Enter the guest's name: ").strip()
            try:
                age = int(input("Enter the guest's age: ").strip())
            except ValueError:
                print("Age must be an integer. Try again.")
                continue
            email = input("Enter the guest's email: ").strip()
            add_guest(name, age, email)
        elif choice == "2":
            name = input("Enter the name of the guest to remove: ").strip()
            remove_guest(name)
        elif choice == "3":
            name = input("Enter the guest's name: ").strip()
            print(get_guest_info(name))
        elif choice == "4":
            display_guests_by_age()
        elif choice == "5":
            print(f"Total number of guests: {guest_count()}")
        elif choice == "6":
            print("Exiting the Guest Manager. Have a great party!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

# Run the guest manager
guest_manager()

````
